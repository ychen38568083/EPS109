```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

$\textbf{PART 1}$

```python
N = 51
L=52
g = np.zeros((L,L))
plt.imshow(g)
for k in range(3):
    for i in range(35):
        g[9 + i][10+k] = 100
        
for j in range(3):
    for k in range(30):
        for i in range(3):
            g[(9 + i)+(j*17)][10+k] = 100
            
%matplotlib qt
plt.imshow(g)

x = np.linspace(0,L,N+1)
y = np.linspace(0,L,N+1)
X, Y = np.meshgrid(x,y)
fig = plt.figure(dpi=200)


#T = np.zeros((L,L))
# for i in range(L):
#     T[i][0] = 100
T = g
Tnew = T
iterations = 100

for k in range(iterations):
    
    if k == 0:
        tot0 = np.sum(T)
        print('Total value initially =', tot0)

    for i in range(1,L-1):
         for j in range(1,L-1):
            dym = T[i-1][j]
            dyp = T[i+1][j]
            dxm = T[i][j-1]
            dxp = T[i][j+1]
            Tnew[i][j] = (dxp + dxm + dyp + dym)/4
        
    if k == iterations-1:
        totf = np.sum(T)
        print('Total value at the end =', totf)
    fig.clear()
    ax = fig.gca(projection='3d')
    ax.set_zlim(top=100)
    ax.plot_surface(X,Y,T,cmap=cm.coolwarm, antialiased=False,vmin=0,vmax=100)
    plt.draw()
    plt.pause(0.03)
```

$\textbf{PART 2}$

```python
N = 51
L=52
g = np.zeros((L,L))
plt.imshow(g)
for k in range(3):
    for i in range(35):
        g[9 + i][10+k] = 100
        
for j in range(3):
    for k in range(30):
        for i in range(3):
            g[(9 + i)+(j*17)][10+k] = 100
            
%matplotlib qt
plt.imshow(g)


x = np.linspace(0,L,N+1)
y = np.linspace(0,L,N+1)
X, Y = np.meshgrid(x,y)
fig = plt.figure(dpi=200)


#T = np.zeros((L,L))
# for i in range(L):
#     T[i][0] = 100
T = g
Tnew = T
iterations = 100

for k in range(iterations):
    
    if k == 0:
        tot0 = np.sum(T)
        print('Total value initially =', tot0)

    for i in range(1,L-1):
         for j in range(1,L-1):
            dym = T[i-1][j]
            dyp = T[i+1][j]
            dxm = T[i][j-1]
            dxp = T[i][j+1]
            Tnew[i][j] = (dxp + dxm + dyp + dym)/4
            
            #average out boundaries for part 2
            if j == 1:
                Tnew[i][j-1] = (Tnew[i][j-1] + Tnew[i][j])/2
                Tnew[i][j] = (Tnew[i][j-1] + Tnew[i][j])/2
                
            if j == L-2:
                Tnew[i][j+1] = (Tnew[i][j+1] + Tnew[i][j])/2
                Tnew[i][j] = (Tnew[i][j+1] + Tnew[i][j])/2
                             
            if i == 1:
                Tnew[i-1][j] = (Tnew[i-1][j] + Tnew[i][j])/2
                Tnew[i][j] = (Tnew[i-1][j] + Tnew[i][j])/2
                
            if i == L-2:
                Tnew[i+1][j] = (Tnew[i+1][j] + Tnew[i][j])/2
                Tnew[i][j] = (Tnew[i+1][j] + Tnew[i][j])/2
            
    if k == iterations-1:
        totf = np.sum(T)
        print('Total value at the end =', totf)
    fig.clear()
    ax = fig.gca(projection='3d')
    ax.set_zlim(top=100)
    ax.plot_surface(X,Y,T,cmap=cm.coolwarm, antialiased=False,vmin=0,vmax=100)
    plt.draw()
    plt.pause(0.03)

print('About',(1-(totf/tot0))*100,'percent of the original value is lost after',iterations,'iterations.')
```
