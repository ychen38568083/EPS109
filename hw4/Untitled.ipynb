```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib tk
```

```python
n = 11
m = 2*np.floor(2*np.random.random(size=(n,n)))-1
print(m)
plt.imshow(m)
plt.show()
```

```python
nBlocks = 1000
Temp = 10
J = -1
for iBlock in range(0,nBlocks):
    for ix in range(0,n):
        ixp = (ix+1) % n
        ixm = (ix-1) % n
        for iy in range(0,n):
            iyp = (iy+1) % n
            iym = (iy-1) % n
            s= m[ixp,iy]+m[ixm,iy]+m[ix,iyp]+m[ix,iym] #sum up all the neighbor spins
            mold = m[ix,iy]*s #multiply by the energy of center spin
            Eold = J*mold*s#old energy
            mNew = -mold
            Enew = J*mNew*s
            Ediff = Enew - Eold
            p = np.exp(-Ediff/Temp) #probability
            if np.random.random() < p: #if random chance is less than the probability then we accept
                m[ix,iy] = mNew
    plt.imshow(m)
    plt.draw()
    plt.pause(0.01)
```
