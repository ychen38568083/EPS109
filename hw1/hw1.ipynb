$\textbf{Question 4}$

```python
plt.figure(figsize=(10,5))
x = np.arange(0,360,2)
y1 = 3*np.cos(x)
y2 = 2*np.cos(2*x)
y3 = np.cos(3*x)
plt.plot(y1, 'r:')
plt.plot(y2, 'b--')
plt.plot(y3, 'g-')
plt.legend(['3cos(x)', '2cos(2x)', 'cos(3x)'], fontsize=15)
plt.xlabel("Periodic Time",fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.grid(True)
plt.show()
```

$\textbf{Question 5}$

```python
x = np.arange(0,1,0.01)
r = np.array([0.5,1.5,2.5,3.5,4])
y = x
plt.figure(figsize=(10,5))
plt.plot(x,y)
for i in np.arange(r.size):
    g = r[i]*x*(1-x)
    plt.plot(x, g,'--')
    
plt.legend(['f(x)=x','r = 0.5','r = 1.5', 'r = 2.5', 'r = 3.5', 'r = 4'],fontsize=15)
plt.grid(True)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x), g(x)',fontsize=16)
plt.show()
```

Only r = [1.5, 2.5, 3.5, 4] will have f(x) and g(x) intersact eachother.

$\textbf{Question 6}$

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
for i in np.arange(r.size):
    g = r[i]*x*(1-x)
    h = r[i]*g*(1-g)
    plt.plot(x,h, '--')

plt.legend(['f(x)=x','r = 0.5','r = 1.5', 'r = 2.5', 'r = 3.5', 'r = 4'],fontsize=15)
plt.grid(True)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x), h(x)',fontsize=16)
plt.show()
plt.show()
```
