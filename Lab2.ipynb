```python
n = 11
data = np.zeros((n,n))
for i in range(0,n):
    data[i,3] = 1
    
col = (n-1)//2
for i in range(0,n,2):
    data[i,col] = 1
    
print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
for i in range(0,n):
    data[i,n-1-i] = 1
    data[i,i] = 1
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))

data[1,1:n-1] = 1
data[n-2,1:n-1] = 1
data[1:9, 1] = 1
data[1:9, n-2] =1
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
for i in range(0,n,2):
    data[i,i:n-1*i] = 1

for i in range(0,n,2):
    data[n-1-i, i:n-1*i] = 1
    
for i in range(0,n,2):
    data[i:n-i-1,i] = 1
    
for i in range(0,n,2):
    data[i:n-i-1,n-1-i] = 1
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))

for i in range(4):
    data[i+1,1:5] = 1
    data[i+6,6:n-1] = 1
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n = 11
data=np.zeros((n,n))

for i in range(4):
    data[i+1,1:5] = 1
    data[i+6,6:n-1] = 1

    data[i+6,1:5] = 1
    data[i+1,6:n-1]=1
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n = 11
data=np.zeros((n,n))

lol = np.arange(2,12,2)
for i in range(5):
    data[i+1:n-(i+1),n-lol[i]]= 1
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n =501
data6 = np.zeros((n,n))
r = n/2.0
center = n//2 #appoint a center (note this is for both x and y components, meaning h = k = n//2)
print(r)
for i in range(n):
    for j in range(n):
        if (i-center)**2+(j-center)**2 <= r**2:
            data6[i,j] = 1
plt.imshow(data6,interpolation='nearest')
plt.show()
```
