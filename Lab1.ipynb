```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline
```

```python
n = 51
L = 10.0
x = np.arange(-L, L, L/n)
y = np.arange(-L, L, L/n)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2+Y**2) + 1e-14
Z = np.sin(R)/R
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0, antialiased=False)
```

```python
data = np.loadtxt('ice_core_temperature_data.txt', usecols=(2,4))
deut = np.loadtxt('ice_core_temperature_data.txt', usecols=(3))
print(data)
```

```python
plt.plot(data[:,0], data[:,1], 'r-', linewidth=1, markersize=0, mec='r',mew=2,mfc='pink', label='sin(x)')
plt.xlabel('Age [Years]')
plt.ylabel('Temperature [K]')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.show()
plt.plot(data[:,0], deut)
plt.xlabel('Age [Years]')
plt.ylabel('Deuterium Level')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.show()
```

```python
temp = data[:,1]
average = sum(temp)/temp.shape[0]
print(average)
plt.plot(data[:,0], data[:,1], 'r-', linewidth=1, markersize=0, mec='r',mew=2,mfc='pink', label='sin(x)')
plt.fill_between(data[:,0], data[:,1],average,color=(1.0,0.0,0.9),lw=0)
plt.xlabel('Years before 1950')
plt.ylabel('temperature - reference temperature [k]')
plt.show()
```
