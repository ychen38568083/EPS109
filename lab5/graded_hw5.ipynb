```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MyFirstFunction(x,r):
    f=r/4.0 * (np.sin(x*np.pi))**2
    return f
```

```python
n = 100
r=3.4
x=0.49
xx = []
yy = []
for i in range(0,n):
    xx.append(i)
    yy.append(x)
    x = MyFirstFunction(x,r)
```

```python
%matplotlib inline
plt.figure(dpi=300)
plt.plot(xx,yy,'ro-',mfc='white')
plt.show()
```

```python
def LM(x1,r1):
    f1 = r1*x1*(1-x1)
    return f1

def LM2(x2,r2):
    #f2 = r2*LM(x2,r2)*(1-LM(x2,r2))
    f2 = LM(LM(x2,r2),r2)
    return f2

def LM3(x3,r3):
    #f3 = r3*LM2(x3,r3)*(1-LM2(x3,r3))
    f3 = LM2(LM2(x3,r3),r3)
    return f3

def LM4(x4,r4):
    #f4 = r4*LM3(x4,r4)*(1-LM3(x4,r4))
    f4 = LM3(LM3(x4,r4),r4)
    return f4
```

```python
def allFunc(x,r):
    plt.plot(x,LM(x,r))
    plt.plot(x,LM2(x,r))
    plt.plot(x,LM3(x,r))
    plt.plot(x,LM4(x,r))
    plt.plot(x,x)
    plt.show()
```

```python
xtest = np.arange(0,1,1/500)
rtest = 3.4
```

```python
allFunc(xtest,rtest)
```
