```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 50
z0 = 0
#c = [-4,-2.001,-2,-1.999,-1,0,0.5,1,2]
c = 2

for i in range(0,n):
    z0 = z0**2 + c
    print(z0)
    if z0 > 50:
        print('Diverges')
        break
```

```python
cGrid = np.linspace(-4,2,50)
for i in range(0,len(cGrid)):
    z = 0
    zList = []
    for k in range(0,len(cGrid)):
        z = z**2 + cGrid[i]
        if z > 50:
            zList.append(1)
        if z < 50:
            zList.append(0)
            
    plt.plot(zList,cGrid)
plt.show()
            
        
```

```python
n = 50
#c = [-4,-2.001,-2,-1.999,-1,0,0.5,1,2]
cList = [complex(-0.4,0.2),
         complex(0.4,0.6),
         complex(-1.7,+0.001),
         -1.7,
         complex(-1.7,0.001),
         -1.778,
         complex(-1.778,0.001),
         -1.25,
         complex(-1.25,0.04),
         complex(-0.125,-0.9),
         complex(-0.125,-0.85)]

for k in range(0,len(cList)):
    print('--------------------------',cList[k],'--------------------------')
    z0 = 0
    for i in range(0,n):
        z0 = z0**2 + cList[k]
        print(z0)
        if abs(z0) > 50:
            print('Diverges')
            break
```

```python
def mandelbrot(xmin,xmax,ymin,ymax,N,n):
    #N =201
    x = np.linspace(xmin,xmax,N)
    y = np.linspace(ymin,ymax,N)
    #n=75
    matrix = np.zeros((N,N))
    for i in range(0,N):
        for j in range(0,N):
            C = complex(x[i],y[j])
            z = 0
            for k in range(0,n):
                z = z**2 + C
                if abs(z) > 50: # if iteration goes over 50, it diverges
                    matrix[j][i] = k
                    break
                else:
                    matrix[j][i] = k
                
                
    plt.figure(figsize=(15,10))           
    plt.imshow(matrix)

    plt.show()
```

```python
mandelbrot(-.8,-.7,0.05,0.15,201,50)
mandelbrot(-.8,-.7,0.05,0.15,201,75)
mandelbrot(-.8,-.7,0.05,0.15,201,100)
```
