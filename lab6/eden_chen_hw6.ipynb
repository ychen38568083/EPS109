```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mandelbrot(xmin,xmax,ymin,ymax,N,n):
    #N =201 #resolution of the matrix
    x = np.linspace(xmin,xmax,N)
    y = np.linspace(ymin,ymax,N)
    #n=75
    matrix = np.zeros((N,N))
    for i in range(0,N):
        for j in range(0,N):
            C = complex(-.4,.6)
            z = complex(x[i],y[j])
            for k in range(0,n):
                z = z**2 + C
                if abs(z) > 50:
                    matrix[j][i] = k
                    break
                else:
                    matrix[j][i] = k
                
                
    plt.figure(figsize=(20,15))           
    plt.imshow(matrix)
    plt.show()
```

```python

mandelbrot(-1.5,1.5,-1.1,1.1,501,100)
```
