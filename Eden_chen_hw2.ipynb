```python
#Extra Credit

#Background 1s
n= 1000
data = np.zeros((n,n))
data[0:n,0:n] = 1

r = n/2.05
center1 = n//2 #appoint a center (note this is for both x and y components, meaning h = k = n//2)
#0 Circle 
for i in range(n):
    for j in range(n):
        if (i-center1)**2+(j-center1)**2 <= r**2:
            data[i,j] = 0

#Inner 1 Circle            
r2 = n/2.15
for i in range(n):
    for j in range(n):
        if (i-center1)**2+(j-center1)**2 <= r2**2:
            data[i,j] = 1
#Mouth
r4 = n//3
for i in range(n):
    for j in range(n):
        if (i-n//2)**2+(j-n//1.9)**2 <= r4**2:
            data[i,j] = 0

#mouth cover
data[150:850,190:650] = 1

#Eyes
r3 = n//12
x_left = n//3.0
y = n//3#same elevation for both eyes
x_right = n//1.5 
#Left Eye
for i in range(n):
    for j in range(n):
        if (i-x_left)**2+(j-y)**2 <= r3**2:
            data[i,j] = 0
#Right Eye
for i in range(n):
    for j in range(n):
        if (i-x_right)**2+(j-y)**2 <= r3**2:
            data[i,j] = 0
#rotate image 3 times            
for i in range(3):
    data = np.rot90(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```
