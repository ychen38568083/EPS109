```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
N = 50
totN = 5*N

T = np.zeros(totN+1)
T[100:150] = 1250

#dt = 10e-4
dx=500/N # 500 is the width in cm of the dike, while the dike has N cells
k=0.025
Cp= 1.2
density = 2.75
r = 0.273
dt = r*((dx**2)*(Cp*density))/k # dt is in second
print('Time of each iteration is',dt,'seconds')

hour = (dt/3600)
print(hour)
print('1 hour takes',hour,'iterations')
day = hour*24
print('1 day takes', day, 'iterations')
month = day*30
print('1 month takes', month, 'iterations')
year = month*365
print('1 year takes', year, 'iterations')

timeframes = [1, 24, 24*30, 24*30*12] #1 hour, 1 day, 1 month, 1 year
time_sec = ['hour','day','month','year']

for k in range(4):
    for j in range(timeframes[k]):
        for i in range(totN):
            T[i] = r *(T[i+1] - 2*T[i] +T[i-1]) + T[i]
            
    

    print('Temp after 1',time_sec[k],'is',T.max(),'kelvin')
    #plt.clf()
    plt.plot(T,'.-',mfc='w')
    plt.ylim(0,1350)
    #plt.draw()
    #plt.pause(.005)

plt.legend(['1 Hour','1 Day','1 Month', '1 Year'])
plt.show()
```
